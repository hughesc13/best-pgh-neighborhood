```python
import matplotlib.pyplot as plt
import pandas as pd

# Load and process data
df = pd.read_csv("Pools.csv")
df['neighborhood'] = df['neighborhood'].str.strip().str.title()

pool_counts = df.groupby('neighborhood').size().reset_index(name='pool_count')
pool_counts['rank'] = pool_counts['pool_count'].rank(method='dense', ascending=False).astype(int)
ranked_table = pool_counts.sort_values(by='rank')

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.barh(ranked_table['neighborhood'], ranked_table['pool_count'], color='skyblue')
plt.xlabel("Number of Pools/Spray Parks")
plt.title("Pools and Spray Parks by Neighborhood in Pittsburgh")
plt.gca().invert_yaxis()  # Rank 1 at the top
plt.tight_layout()
plt.show()
```
